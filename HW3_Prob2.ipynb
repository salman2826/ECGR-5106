```python
import time
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import numpy as np
import requests
import matplotlib.pyplot as plt

# -------------------------------
# 1. Download and Preprocess Data
# -------------------------------
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
response = requests.get(url)
text = response.text  # Entire Tiny Shakespeare text

# Create character mappings
chars = sorted(list(set(text)))
vocab_size = len(chars)
char_to_int = {ch: i for i, ch in enumerate(chars)}
int_to_char = {i: ch for i, ch in enumerate(chars)}

# Create a dataset given a sequence length
def create_dataset(text, seq_length):
    encoded_text = [char_to_int[ch] for ch in text]
    sequences = []
    targets = []
    for i in range(0, len(encoded_text) - seq_length):
        seq = encoded_text[i:i+seq_length]
        target = encoded_text[i+seq_length]
        sequences.append(seq)
        targets.append(target)
    sequences = torch.tensor(sequences, dtype=torch.long)
    targets = torch.tensor(targets, dtype=torch.long)
    return sequences, targets

# -------------------------------
# 2. Dataset and DataLoader
# -------------------------------
class CharDataset(Dataset):
    def __init__(self, sequences, targets):
        self.sequences = sequences
        self.targets = targets

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, index):
        return self.sequences[index], self.targets[index]

def get_dataloaders(seq_length, batch_size=128, train_split=0.8):
    sequences, targets = create_dataset(text, seq_length)
    dataset = CharDataset(sequences, targets)
    train_size = int(len(dataset) * train_split)
    test_size = len(dataset) - train_size
    train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])
    train_loader = DataLoader(train_dataset, shuffle=True, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, shuffle=False, batch_size=batch_size)
    return train_loader, test_loader

# -------------------------------
# 3. Model Definitions (LSTM and GRU)
# -------------------------------
class CharRNNModel(nn.Module):
    def __init__(self, model_type, vocab_size, embedding_dim, hidden_dim, num_layers, fc_hidden_dim=None):
        """
        model_type: 'LSTM' or 'GRU'
        fc_hidden_dim: if provided, an extra fully connected layer is applied before the final output.
        """
        super(CharRNNModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.model_type = model_type
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        elif model_type == 'GRU':
            self.rnn = nn.GRU(embedding_dim, hidden_dim, num_layers, batch_first=True)
        else:
            raise ValueError("model_type must be 'LSTM' or 'GRU'")
        
        
        if fc_hidden_dim is not None:
            self.fc1 = nn.Linear(hidden_dim, fc_hidden_dim)
            self.fc2 = nn.Linear(fc_hidden_dim, vocab_size)
        else:
            self.fc1 = None
            self.fc2 = nn.Linear(hidden_dim, vocab_size)
    
    def forward(self, x, hidden=None):
        # x shape: (batch, seq_length)
        embedded = self.embedding(x)  # (batch, seq_length, embedding_dim)
        out, hidden = self.rnn(embedded, hidden)
        
        last_output = out[:, -1, :]  # (batch, hidden_dim)
        if self.fc1 is not None:
            x = torch.relu(self.fc1(last_output))
            logits = self.fc2(x)
        else:
            logits = self.fc2(last_output)
        return logits, hidden

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# -------------------------------
# 4. Training and Evaluation Functions
# -------------------------------
def train_model(model, train_loader, test_loader, epochs, device, lr=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.to(device)
    
    train_losses = []
    test_losses = []
    test_accuracies = []
    
    start_time = time.time()
    for epoch in range(1, epochs+1):
        model.train()
        epoch_loss = 0
        for x, y in train_loader:
            x = x.to(device)
            y = y.to(device)
            optimizer.zero_grad()
            logits, _ = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
        avg_train_loss = epoch_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        
        # Evaluate on test set
        model.eval()
        total_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for x, y in test_loader:
                x = x.to(device)
                y = y.to(device)
                logits, _ = model(x)
                loss = criterion(logits, y)
                total_loss += loss.item()
                preds = torch.argmax(logits, dim=1)
                correct += (preds == y).sum().item()
                total += y.size(0)
        avg_test_loss = total_loss / len(test_loader)
        test_losses.append(avg_test_loss)
        accuracy = correct / total
        test_accuracies.append(accuracy)
        print(f"Epoch {epoch}/{epochs} | Train Loss: {avg_train_loss:.4f} | Test Loss: {avg_test_loss:.4f} | Test Acc: {accuracy:.4f}")
    training_time = time.time() - start_time
    return train_losses, test_losses, test_accuracies, training_time

# -------------------------------
# 5. Plotting Functions
# -------------------------------
def plot_metrics(epochs, train_losses, test_losses, test_accuracies, model_type, seq_length):
    epoch_range = range(1, epochs+1)
    
    # Plot Loss: Training and Validation Loss vs. Epoch
    plt.figure()
    plt.plot(epoch_range, train_losses, label='Training Loss')
    plt.plot(epoch_range, test_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{model_type} - Seq Length {seq_length} Loss')
    plt.legend()
    plt.show()
    
    # Plot Accuracy: Validation Accuracy vs. Epoch
    plt.figure()
    plt.plot(epoch_range, test_accuracies, label='Validation Accuracy', color='green')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{model_type} - Seq Length {seq_length} Accuracy')
    plt.legend()
    plt.show()

# -------------------------------
# 6. Sampling from the Trained Model
# -------------------------------
def sample_text(model, seed, length, device):
    model.eval()
    
    indices = [char_to_int[ch] for ch in seed]
    input_seq = torch.tensor(indices, dtype=torch.long).unsqueeze(0).to(device)
    generated = seed
    hidden = None
    for _ in range(length):
        logits, hidden = model(input_seq, hidden)
        probs = torch.softmax(logits, dim=1)
        
        next_idx = torch.multinomial(probs, num_samples=1).item()
        next_char = int_to_char[next_idx]
        generated += next_char
        
        new_seq = input_seq.squeeze(0).tolist()[1:] + [next_idx]
        input_seq = torch.tensor(new_seq, dtype=torch.long).unsqueeze(0).to(device)
    return generated

# -------------------------------
# 7. Experiment Function
# -------------------------------
def run_experiment(seq_length, model_type, config, epochs=10):
    print(f"\n=== Experiment: Model: {model_type} | Sequence Length: {seq_length} | Config: {config} ===")
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    train_loader, test_loader = get_dataloaders(seq_length, batch_size=config['batch_size'])
    model = CharRNNModel(
        model_type, 
        vocab_size, 
        embedding_dim=config['embedding_dim'], 
        hidden_dim=config['hidden_dim'], 
        num_layers=config['num_layers'], 
        fc_hidden_dim=config.get('fc_hidden_dim', None)
    )
    num_params = count_parameters(model)
    print(f"Model parameters: {num_params}")
    
    train_losses, test_losses, test_accuracies, training_time = train_model(
        model, train_loader, test_loader, epochs, device, lr=config['lr']
    )
    print(f"Total training time: {training_time:.2f} seconds")
    
    
    seed = text[:seq_length]
    generated_text = sample_text(model, seed, 100, device)
    print("Sample generated text:")
    print(generated_text)
    
    # Plot the training and validation metrics
    plot_metrics(epochs, train_losses, test_losses, test_accuracies, model_type, seq_length)
    
    return {
        'model_type': model_type,
        'seq_length': seq_length,
        'num_params': num_params,
        'train_losses': train_losses,
        'test_losses': test_losses,
        'test_accuracies': test_accuracies,
        'training_time': training_time,
        'sample_text': generated_text
    }

# -------------------------------
# 8. Main: Run Experiments and Compare Results
# -------------------------------
def main():
    # Hyperparameter configuration
    config = {
        'embedding_dim': 128,
        'hidden_dim': 256,
        'num_layers': 2,
        'fc_hidden_dim': 128,  
        'batch_size': 128,
        'lr': 0.001
    }
    
    # Sequence lengths to experiment with: 20, 30, and 50
    seq_lengths = [20, 30, 50]
    model_types = ['LSTM', 'GRU']
    results = []
    epochs = 10  
    
    for seq_length in seq_lengths:
        for model_type in model_types:
            res = run_experiment(seq_length, model_type, config, epochs=epochs)
            results.append(res)
    
    # Summary of experiments
    print("\n--- Experiment Summary ---")
    for res in results:
        print(f"Model: {res['model_type']} | Seq Length: {res['seq_length']} | Params: {res['num_params']} | "
              f"Final Test Acc: {res['test_accuracies'][-1]:.4f} | Training Time: {res['training_time']:.2f} sec")
    
if __name__ == "__main__":
    main()
```

```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import requests
import matplotlib.pyplot as plt

# -------------------------------
# 1. Download and Preprocess Data
# -------------------------------
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
response = requests.get(url)
text = response.text  

# Create character mappings
chars = sorted(list(set(text)))
vocab_size = len(chars)
char_to_int = {ch: i for i, ch in enumerate(chars)}
int_to_char = {i: ch for i, ch in enumerate(chars)}

def encode(text):
    return [char_to_int[ch] for ch in text]

# -------------------------------
# 2. Create Dataset and DataLoader for Sequence Length 30
# -------------------------------
def create_dataset(text, seq_length):
    encoded = encode(text)
    sequences = []
    targets = []
    for i in range(len(encoded) - seq_length):
        sequences.append(encoded[i:i+seq_length])
        targets.append(encoded[i+seq_length])
    sequences = torch.tensor(sequences, dtype=torch.long)
    targets = torch.tensor(targets, dtype=torch.long)
    return sequences, targets

class CharDataset(Dataset):
    def __init__(self, sequences, targets):
        self.sequences = sequences
        self.targets = targets

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        return self.sequences[idx], self.targets[idx]

def get_dataloaders(seq_length=30, batch_size=128, train_split=0.8):
    sequences, targets = create_dataset(text, seq_length)
    dataset = CharDataset(sequences, targets)
    train_size = int(len(dataset) * train_split)
    test_size = len(dataset) - train_size
    train_ds, test_ds = torch.utils.data.random_split(dataset, [train_size, test_size])
    train_loader = DataLoader(train_ds, shuffle=True, batch_size=batch_size)
    test_loader = DataLoader(test_ds, shuffle=False, batch_size=batch_size)
    return train_loader, test_loader

# -------------------------------
# 3. Model Definition: LSTM/GRU with Configurable Hyperparameters
# -------------------------------
class CharRNNModel(nn.Module):
    def __init__(self, model_type, vocab_size, embedding_dim, hidden_dim, num_layers, fc_hidden_dim=None):
        """
        model_type: 'LSTM' or 'GRU'
        fc_hidden_dim: If provided, adds an extra fully-connected layer before the final output.
        """
        super(CharRNNModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.model_type = model_type
        
        if model_type == 'LSTM':
            self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        elif model_type == 'GRU':
            self.rnn = nn.GRU(embedding_dim, hidden_dim, num_layers, batch_first=True)
        else:
            raise ValueError("model_type must be 'LSTM' or 'GRU'")
        
        # Fully-connected layers
        if fc_hidden_dim is not None:
            self.fc1 = nn.Linear(hidden_dim, fc_hidden_dim)
            self.fc2 = nn.Linear(fc_hidden_dim, vocab_size)
        else:
            self.fc1 = None
            self.fc2 = nn.Linear(hidden_dim, vocab_size)
    
    def forward(self, x, hidden=None):
        # x shape: (batch, seq_length)
        embed = self.embedding(x)  # (batch, seq_length, embedding_dim)
        out, hidden = self.rnn(embed, hidden)
        # Use the last time step’s output
        last_out = out[:, -1, :]  # (batch, hidden_dim)
        if self.fc1 is not None:
            x = torch.relu(self.fc1(last_out))
            logits = self.fc2(x)
        else:
            logits = self.fc2(last_out)
        return logits, hidden

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# -------------------------------
# 4. Training and Evaluation Functions
# -------------------------------
def train_model(model, train_loader, test_loader, epochs, device, lr=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.to(device)
    
    train_losses = []
    test_losses = []
    test_accuracies = []
    
    train_start = time.time()
    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            logits, _ = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        avg_train_loss = running_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        
        # Evaluate on validation set
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for x, y in test_loader:
                x, y = x.to(device), y.to(device)
                logits, _ = model(x)
                loss = criterion(logits, y)
                val_loss += loss.item()
                preds = torch.argmax(logits, dim=1)
                correct += (preds == y).sum().item()
                total += y.size(0)
        avg_val_loss = val_loss / len(test_loader)
        test_losses.append(avg_val_loss)
        accuracy = correct / total
        test_accuracies.append(accuracy)
        print(f"Epoch {epoch}/{epochs} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Val Acc: {accuracy:.4f}")
    
    training_time = time.time() - train_start
    return train_losses, test_losses, test_accuracies, training_time

def sample_text(model, seed, length, device):
    model.eval()
    indices = [char_to_int[ch] for ch in seed]
    input_seq = torch.tensor(indices, dtype=torch.long).unsqueeze(0).to(device)
    generated = seed
    hidden = None
    # Measure inference time
    infer_start = time.time()
    for _ in range(length):
        logits, hidden = model(input_seq, hidden)
        probs = torch.softmax(logits, dim=1)
        next_idx = torch.multinomial(probs, num_samples=1).item()
        generated += int_to_char[next_idx]
        new_seq = input_seq.squeeze(0).tolist()[1:] + [next_idx]
        input_seq = torch.tensor(new_seq, dtype=torch.long).unsqueeze(0).to(device)
    inference_time = time.time() - infer_start
    return generated, inference_time

def plot_metrics(epochs, train_losses, val_losses, test_accuracies, model_type):
    epoch_range = range(1, epochs+1)
    # Plot Losses
    plt.figure()
    plt.plot(epoch_range, train_losses, label='Train Loss')
    plt.plot(epoch_range, val_losses, label='Validation Loss')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"{model_type} Loss Curve (Seq Length 30)")
    plt.legend()
    plt.show()
    
    # Plot Accuracy
    plt.figure()
    plt.plot(epoch_range, test_accuracies, label='Validation Accuracy', color='green')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(f"{model_type} Accuracy Curve (Seq Length 30)")
    plt.legend()
    plt.show()

# -------------------------------
# 5. Experiment: Run for Sequence Length 30
# -------------------------------
def run_experiment(model_type, config, epochs=10):
    print(f"\n=== Running {model_type} Experiment (Seq Length = 30) ===")
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    train_loader, test_loader = get_dataloaders(seq_length=30, batch_size=config['batch_size'])
    
    model = CharRNNModel(
        model_type,
        vocab_size,
        embedding_dim=config['embedding_dim'],
        hidden_dim=config['hidden_dim'],
        num_layers=config['num_layers'],
        fc_hidden_dim=config.get('fc_hidden_dim', None)
    )
    num_params = count_parameters(model)
    print(f"Model Parameters: {num_params}")
    
    # Train the model
    train_losses, val_losses, val_accuracies, training_time = train_model(model, train_loader, test_loader, epochs, device, lr=config['lr'])
    print(f"Total Training Time: {training_time:.2f} sec")
    
    # Plot the training and validation curves
    plot_metrics(epochs, train_losses, val_losses, val_accuracies, model_type)
    
    # Inference
    seed = text[:30]
    output_sequence, inference_time = sample_text(model, seed, length=100, device=device)
    print(f"Inference Time: {inference_time:.4f} sec")
    print("Output Sequence:")
    print(output_sequence)
    
    return {
        'model_type': model_type,
        'num_params': num_params,
        'training_time': training_time,
        'inference_time': inference_time,
        'final_train_loss': train_losses[-1],
        'final_val_loss': val_losses[-1],
        'final_val_acc': val_accuracies[-1],
        'output_sequence': output_sequence
    }

# -------------------------------
# 6. Main: Change Hyperparameters and Run the Experiment for Seq Length 30
# -------------------------------
def main():
    # Change hyperparameters here:
    config = {
        'embedding_dim': 128,
        'hidden_dim': 512,      # Increase number of hidden states
        'num_layers': 3,        # Increase number of hidden layers
        'fc_hidden_dim': 256,   # Fully connected layer dimension
        'batch_size': 128,
        'lr': 0.001
    }
    epochs = 10  

    # Run experiments for LSTM and GRU with sequence length 30
    results_lstm = run_experiment("LSTM", config, epochs)
    results_gru  = run_experiment("GRU", config, epochs)
    
    # Print summary of results
    print("\n--- Summary ---")
    for res in [results_lstm, results_gru]:
        print(f"Model: {res['model_type']}")
        print(f"  Parameters       : {res['num_params']}")
        print(f"  Final Train Loss : {res['final_train_loss']:.4f}")
        print(f"  Final Val Loss   : {res['final_val_loss']:.4f}")
        print(f"  Final Val Acc    : {res['final_val_acc']:.4f}")
        print(f"  Training Time    : {res['training_time']:.2f} sec")
        print(f"  Inference Time   : {res['inference_time']:.4f} sec")
        print(f"  Output Sequence  : {res['output_sequence'][:100]}...\n")

if __name__ == "__main__":
    main()
```
